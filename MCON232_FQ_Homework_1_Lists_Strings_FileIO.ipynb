### Task 2A â€” Count & filter (no comprehensions)

Given a list of numbers:
1) Count how many are **>= 10**
2) Build a new list containing only numbers **>= 10**


````python
nums = [3, 8, 2, 11, 5, 10, 10, 1]

# Required:
# count_big  -> how many numbers >= 10
# big_nums   -> list of numbers >= 10

count_big = 0
big_nums = []

# TODO: write your loop here
for i in range(len(nums)):
  if nums[i] >= 10:
    big_nums.append(nums[i])
    count_big += 1

print("count_big:", count_big)
print("big_nums:", big_nums)
````

### Task 2B â€” Two accumulators

Compute:
1) `total_sum` of the list
2) `even_product` = product of ONLY the even numbers  
   - Start `even_product` at 1  
   - Multiply only when the number is even


````python
nums = [4, 7, 2, 9, 10, 3, 8]

total_sum = 0
even_product = 1

# TODO: write your loop here
for i in range(len(nums)):
  total_sum += nums[i]
  if nums[i] % 2 == 0:
    even_product *= nums[i]

print("total_sum:", total_sum)
print("even_product:", even_product)
````

Find the index of the **first** `"hello"` using a loop.

## Part 3 â€” File I/O Refresher (Colab-Friendly) (25 minutes)

### Step 3A â€” Create the input file (provided)

Run this cell **as-is** (do not change it).


````python
# DO NOT CHANGE THIS CELL
sample_text = """Alice,90
Ben,72
Cara,88
Dan,95
Eva,72
"""

with open("grades.txt", "w") as f:
    f.write(sample_text)

print("grades.txt created âœ…")
````

### Step 3B â€” Read and parse the file into lists

Build two **parallel lists**:
- `names`  (strings)
- `scores` (ints)

Requirements:
- Use a loop to read each line
- Use `strip()` to remove the newline
- Use `split(",")` to separate name and score
- Convert score to `int`

***NOTE*** if you are not certain how the string methods work, please look up on the internet.


````python
names = []
scores = []

# TODO:
# 1) Open grades.txt for reading
# 2) Loop through each line
# 3) Strip newline
# 4) Split by comma into name and score
# 5) Append name to names
# 6) Convert score to int and append to scores

# Write your code below:
gradefile = open("grades.txt", "r")

line = gradefile.readline()
while line != "":
  cleaned = line.strip("\n")
  splitline = cleaned.split(",")
  name = splitline[0]
  score = splitline[1]

  names.append(name)
  scores.append(int(score))

  line = gradefile.readline()


print("names:", names)
print("scores:", scores)
````

### Step 3C â€” Analyze the data (thinking)

Using `names` and `scores`, compute:

1) `avg_score` (average score as a float)
2) `highest_score` and the `highest_name` who earned it
3) `count_72` = how many students earned exactly 72
4) `passing_names` = list of students with score **>= 75**

**Do not use list comprehensions.**


````python
# Required variables:
avg_score = 0.0
highest_score = -1
highest_name = ""
count_72 = 0
passing_names = []

# TODO:
# - compute total
# - compute avg_score
# - find highest score + name
# - count 72s
# - build passing_names

# Note from Shaked: I put them all in separate loops because it helps me stay more organized, but I can try to condense it into one loop if you prefer.
total = 0
for i in range(len(scores)):
  total += scores[i]

avg_score = total / len(scores)

for i in range(len(scores)):
  if scores[i] > highest_score:
    highest_score = scores[i]
    highest_name = names[i]

for i in range(len(scores)):
  if scores[i] == 72:
    count_72 += 1

for i in range(len(scores)):
  if scores[i] >= 75:
    passing_names.append(names[i])

print("avg_score:", avg_score)
print("highest:", highest_name, highest_score)
print("count_72:", count_72)
print("passing_names:", passing_names)
````

## Part 4 â€” Remove Duplicates (20 minutes)

You are not using sets yet, so weâ€™ll solve this with a loop.

**Goal:** Build `unique_nums` containing each number only once, in the order it first appears.

Example:
- Input:  `[2, 5, 2, 9, 5, 1]`
- Output: `[2, 5, 9, 1]`

Hint:  
Use this pattern:
- for each `n` in `nums`  
  - if `n` is **not** already in `unique_nums`, append it


````python
nums = [2, 5, 2, 9, 5, 1, 9, 9, 3]

unique_nums = []

# TODO: write your loop here
for n in nums:
  if n not in unique_nums:
    unique_nums.append(n)

print("unique_nums:", unique_nums)
````

Compute `duplicates_removed`.

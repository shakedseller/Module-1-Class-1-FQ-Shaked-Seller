### üß∫ Lists (1D)
Remember:
- A list is just a collection of items in order.
- You can:
  - find its length with `len(my_list)`
  - access items using indexes: `my_list[0]`, `my_list[-1]`
  - change items by assignment: `my_list[2] = "new value"`
- Common list tools:
  - `append(x)` ‚Üí add to the end
  - `insert(i, x)` ‚Üí add at a position
  - `pop()` ‚Üí remove and return an item
  - `remove(x)` ‚Üí remove a specific value

When in doubt: **print the list** and look at it.

### üßµ Strings
Strings are just text ‚Äî and text often needs cleaning.

Two very helpful tools:

Example:
```python

```
